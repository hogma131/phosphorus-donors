```python
pulse_verts = [0.02, 0.02, 0, -0.9, -0.9, 0, 0]
pulse_timings = [1e-9, 1e-9, 2e-9, 20e-9, 1e-9, 1e-9]
time_step = 0.01e-9
lePulse = PulseSequence(pulse_verts, pulse_timings, time_step)
# pulse_verts = [0.02, -0.8]
# pulse_timings = [0.5e-9]
# time_step = 0.1e-12
# lePulse = PulseSequence(pulse_verts, pulse_timings, time_step)
```

```python
lePulse.plot_waveform()
```

```python
len(lePulse.time_vec)
```

```python
output = dev104.do_pulse_sequence(lePulse, psi0=dev104.S20)
```

```python
proj_ST, proj_UD = dev104.extract_projections_from_sim_data(output.states)
```

```python
dev104.plot_pulse_projections(lePulse, proj_ST, proj_UD)
```

```python
out, times = dev104.do_detuning_pulse(0.05,-0.05,psi0=dev104.S20,ramp_time=1.0e-6)
```

```python
proj_ST, proj_UD = dev104.extract_projections_from_sim_data(out.states)
```

```python
plt.figure(1)
plt.plot(times, proj_ST)
plt.legend(['S11 proj', 'T- proj', 'T0 proj', 'T+ proj', 'S20 proj'])
plt.show(block=False)
```
