```python
import matplotlib.pyplot as plt
from scipy.integrate import ode
from scipy.integrate import odeint
import numpy as np
import scipy.constants as const
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
def delta_E(t, alpha=0.3, vDC=0, vAC=0.005, omega=2*np.pi*228.3e6):
    dE = (1-alpha)*const.eV*(vDC+vAC*np.cos(omega*t))
    return dE
```

```python
def dPdt(t, P, gamma0=2*np.pi*1e9, T=0.1):
    '''
    '''
    kb = 1.38e-23
    dPdt = (2*np.pi*1e9)/(1+np.exp(delta_E(t)/(kb*T))) - gamma0*P
    return dPdt
```

```python
def dPdt_odeint(P, t):
    '''
    '''
    kb = 1.38e-23
    gamma0 = 2*np.pi*1e9
    T = 0.1
    dPdt = (2*np.pi*1e9)/(1+np.exp(delta_E(t)/(kb*T))) - gamma0*P
    return dPdt
```

```python
delta_E(0)/const.eV
```

```python
t = np.linspace(0,100e-9,1000)
P0 = 1.0
tf = 100e-9
dt = 0.1e-9
```

```python
test = ode(dPdt).set_integrator('dopri5', method='bdf')
test.set_initial_value(P0)
```

```python
output = []
while test.successful() and test.t < tf:
    test.integrate(test.t+dt)
    output.append(test.y)
    print(test.y)
#     print("%g %g" % (test.t, test.y))
```

```python
plt.plot(output)
```

Using the odeint function seems much easier?

```python
t = np.linspace(0,100e-9,1000)
P0 = 0.0  # the initial condition
ys = odeint(dPdt_odeint, P0, t)
ys = np.array(ys).flatten()
```

```python
plt.figure(1)
plt.plot(t,ys)
plt.figure(2)
plt.plot(t, np.gradient(ys))
```
